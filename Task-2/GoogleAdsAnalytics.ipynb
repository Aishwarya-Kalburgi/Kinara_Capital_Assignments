```python
df = pd.read_excel("Task 2 - Google Ads Analytics.xlsx")
df.head()
```

```python
corr = df.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot = True, cmap="YlGn")
```

1. In this data bounce_rate and average_cpv features are not contributing to the performance of google ads
2. The following features are highly correlated
   1. average_cost and average_cpc
   2. clicks and impressions
   3. interactions and impressions
   4. engagement_rate and average_cpe
   5. clicks and conversions
   6. clicks and interactions
   7. average_cpm and ctr
   8. average_cost and average_cpm
3. As impressions increase clicks also increase and there by conversions and interactions 
4. conversions and clicks for particular ad groups are more compared to other ad groups
